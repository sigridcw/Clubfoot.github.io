```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

Four_mm_XY_Flat_df=pd.read_csv("./Downloads/XY-4mm-flat - XY-4mm-flat.csv")
Four_mm_YZ_Flat_df=pd.read_csv("./Downloads/YZ-4mm-flat - YZ-4mm-flat.csv")
Four_mm_POP_Flat_df=pd.read_csv("./Downloads/POP-4mm-flat - POP-4mm-flat.csv")
Four_mm_XY_Edge_df=pd.read_csv("./Downloads/XY-4mm-edge - XY-4mm-edge.csv")
Four_mm_YZ_Edge_df=pd.read_csv("./Downloads/YZ-4mm-edge - YZ-4mm-edge.csv")
Four_mm_POP_Edge_df=pd.read_csv("./Downloads/POP-4mm-edge - POP-4mm-edge.csv")
```

```python
Strain_Four_mm_XY_Flat_df = Four_mm_XY_Flat_df['Strain']
Stress_Four_mm_XY_Flat_df = Four_mm_XY_Flat_df['Stress']
Load_Four_mm_XY_Flat_df = Four_mm_XY_Flat_df['Load']
Deflection_Four_mm_XY_Flat_df = Four_mm_XY_Flat_df['Deflection']

plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
plt.plot(Deflection_Four_mm_XY_Flat_df, Load_Four_mm_XY_Flat_df)
plt.xlabel('deflection (mm)')
plt.ylabel('load (N)')
plt.title('Load Deflection Curve of ABS 4mm Size in\n4-point bending flatwise test among XY alignment')
plt.legend(['4mm_XY_Flat'])

plt.subplot(1,2,2)
plt.plot(Strain_Four_mm_XY_Flat_df, Stress_Four_mm_XY_Flat_df)
plt.xlabel('strain (%)')
plt.ylabel('stress (MPa)')
plt.title('Stress Strain Curve of ABS 4mm Size in\n4-point bending flatwise test among XY alignment')
plt.legend(['4mm_XY_Flat'])

plt.tight_layout()
plt.savefig('./Downloads/4mm_XY_Flat.png')
plt.show()
```
