```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

Three_mm_XY_Flat_df=pd.read_csv("./Downloads/XY-3mm-flat - XY-3mm-flat.csv")
Three_mm_YZ_Flat_df=pd.read_csv("./Downloads/YZ-3mm-flat - YZ-3mm-flat.csv")
Three_mm_POP_Flat_df=pd.read_csv("./Downloads/POP-3mm-flat - POP-3mm-flat.csv")
Three_mm_XY_Edge_df=pd.read_csv("./Downloads/XY-3mm-edge - XY-3mm-edge.csv")
Three_mm_YZ_Edge_df=pd.read_csv("./Downloads/YZ-3mm-edge - YZ-3mm-edge.csv")
Three_mm_POP_Edge_df=pd.read_csv("./Downloads/POP-3mm-edge - POP-3mm-edge.csv")
```

```python
Strain_Three_mm_XY_Flat_df = Three_mm_XY_Flat_df['Strain']
Stress_Three_mm_XY_Flat_df = Three_mm_XY_Flat_df['Stress']
Load_Three_mm_XY_Flat_df = Three_mm_XY_Flat_df['Load']
Deflection_Three_mm_XY_Flat_df = Three_mm_XY_Flat_df['Deflection']

plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
plt.plot(Deflection_Three_mm_XY_Flat_df, Load_Three_mm_XY_Flat_df)
plt.xlabel('deflection (mm)')
plt.ylabel('load (N)')
plt.title('Load Deflection Curve of ABS 3mm Size in\n4-point bending flatwise test among XY alignment')
plt.legend(['3mm_XY_Flat'])

plt.subplot(1,2,2)
plt.plot(Strain_Three_mm_XY_Flat_df, Stress_Three_mm_XY_Flat_df)
plt.xlabel('strain (%)')
plt.ylabel('stress (MPa)')
plt.title('Stress Strain Curve of ABS 3mm Size in\n4-point bending flatwise test among XY alignment')
plt.legend(['3mm_XY_Flat'])

plt.tight_layout()
plt.savefig('./Downloads/3mm_XY_Flat.png')
plt.show()
```
