```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

Five_mm_XY_Flat_df=pd.read_csv("./Downloads/XY-5mm-flat - XY-5mm-flat.csv")
Five_mm_YZ_Flat_df=pd.read_csv("./Downloads/YZ-5mm-flat - YZ-5mm-flat.csv")
Five_mm_POP_Flat_df=pd.read_csv("./Downloads/POP-5mm-flat - POP-5mm-flat.csv")
Five_mm_XY_Edge_df=pd.read_csv("./Downloads/XY-5mm-edge - XY-5mm-edge.csv")
Five_mm_YZ_Edge_df=pd.read_csv("./Downloads/YZ-5mm-edge - YZ-5mm-edge.csv")
Five_mm_POP_Edge_df=pd.read_csv("./Downloads/POP-5mm-edge - POP-5mm-edge.csv")
```

```python
Strain_Five_mm_XY_Flat_df = Five_mm_XY_Flat_df['Strain']
Stress_Five_mm_XY_Flat_df = Five_mm_XY_Flat_df['Stress']
Load_Five_mm_XY_Flat_df = Five_mm_XY_Flat_df['Load']
Deflection_Five_mm_XY_Flat_df = Five_mm_XY_Flat_df['Deflection']

plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
plt.plot(Deflection_Five_mm_XY_Flat_df, Load_Five_mm_XY_Flat_df)
plt.xlabel('deflection (mm)')
plt.ylabel('load (N)')
plt.title('Load Deflection Curve of ABS 5mm Size in\n4-point bending flatwise test among XY alignment')
plt.legend(['5mm_XY_Flat'])

plt.subplot(1,2,2)
plt.plot(Strain_Five_mm_XY_Flat_df, Stress_Five_mm_XY_Flat_df)
plt.xlabel('strain (%)')
plt.ylabel('stress (MPa)')
plt.title('Stress Strain Curve of ABS 5mm Size in\n4-point bending flatwise test among XY alignment')
plt.legend(['5mm_XY_Flat'])

plt.tight_layout()
plt.savefig('./Downloads/5mm_XY_Flat.png')
plt.show()
```
