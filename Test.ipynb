# Introduction of Clubfoot

> Biology

Clubfoot is a developmental deformation like developmental hip dysplasia and idiopathic scoliosis.
A normally developing foot turns into a clubfoot during the second trimester of pregnancy, and detecting clubfoot with ultrasonography before the 16th week of gestation is rare.

The shape of the tarsal joints is altered.
The forefoot is in pronation with concave plantar arch (cavus).
Increasing flexion of the metatarsal bones is present in a lateromedial direction.

Tibialis posterior, gastrosoleus, and long toe flexors are smaller and shorter in clubfoot comparing with those in normal foot.
The thick and taut ligaments, composed of excessive collagen fibers, of posterior and medial aspect of ankle and tarsal joints severely restrain the foot in equinus, the navicular and calcaneus in adduction and inversion.

Microscopically, the crimp within the connective tissue allows the ligaments to be stretched, which makes manual correction feasible.

> Kinematics

At birth, the tarsal bones are in severe flexion, adduction, and inversion position.
* Talus - plantarflexion
  1. Neck - medially and plantarly deflected
  2. Head - wedge-shaped
* Navicular - medial displacement, close to medial malleolus
* Calcaneus - adduction and inversion under Talus

![Image](https://upload.cc/i1/2021/01/19/Ao89IL.png)

By gradually stretching the taut tarsal ligaments, we can correct the severe displacements of the tarsal bones in clubfoot.
We abduct the foot in supination while applying counterpressure over the lateral aspect of talus head to correct heel varus deformity.
A well-molded cast maintains the foot in desired degree of correction and position to gently stretch the ligaments.
It usually takes five to six cast changes for correction.

The deformity of clubfoot could be categorized into 4 parts:
* Midfoot cavus
* Forefoot adduction
* Hindfoot varus
* Ankle equinus

## Ponseti Management

*`Concept`*

Compare the normal relationships of the tarsal bones [left] with that of clubfoot [right].
Note that:
* Talus (red) is deformed and the navicular (yellow) is medially displaced.
* The foot is rotated around the head of the talus (blue arrow).

![Image](https://upload.cc/i1/2021/01/19/CIvjlt.png)

 Ponseti correction is achieved by reversing this rotation.
 Correction is achieved gradually by serial casts.
 The Ponseti technique corrects the deformity by gradually rotating the foot around the head of the talus (red circle) over a period of weeks during cast correction.

![Image](https://upload.cc/i1/2021/01/19/RcTaKW.png)

*`Timepoint`*

When possible, start soon after birth (7 to 10 days).
However, most clubfoot deformities can be corrected throughout childhood using Ponseti management.

*`Duration`*

It takes approximately **6** weeks by **weekly** manipulations followed by plaster cast applications.
If the deformity is not corrected after 6 or 7 plaster cast changes, the treatment is most likely faulty.

*`Success Rate`*

Mostly > 95%. It depends on degree of stiffness of the foot, experience of physician, and reliability of family.

*`Steps`*
1. ~ 5 casting periods
 * Correct cavus, adductus and varus
2. Percutaneous heel-cord tenotomy
 * Correct equinus completely
3. Place foot in the last cast for 3 weeks
 * Night splint with a foot abduction brace, continued until ~ 2-4 y/o

![Image](https://i.ibb.co/Z62RS31/2021-03-18-1-58-00.png)

*`Challenges`*
1. Training course of physicians: Practice, Knowledge, Skills
2. Plaster splint - Too loose or short
3. Plaster splint - Too tense
4. Plaster splint - Wrong orientation
5. Plaster splint - Border induced dermatological lesions
6. Time-consuming and tough for parents to execute several tasks

2. *Compressive Strain*
 * **Ratio** of compressive deformation to the gage length of test specimen
