```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

Two_mm_XY_Flat_df=pd.read_csv("./Downloads/XY-2mm-flat - XY-2mm-flat.csv")
Two_mm_YZ_Flat_df=pd.read_csv("./Downloads/YZ-2mm-flat - YZ-2mm-flat.csv")
Two_mm_POP_Flat_df=pd.read_csv("./Downloads/POP-2mm-flat - POP-2mm-flat.csv")
Two_mm_YZ_Edge_df=pd.read_csv("./Downloads/YZ-2mm-edge - YZ-2mm-edge.csv")
Two_mm_POP_Edge_df=pd.read_csv("./Downloads/POP-2mm-edge - POP-2mm-edge.csv")
```

```python
Strain_Two_mm_XY_Flat_df = Two_mm_XY_Flat_df['Strain']
Stress_Two_mm_XY_Flat_df = Two_mm_XY_Flat_df['Stress']
Load_Two_mm_XY_Flat_df = Two_mm_XY_Flat_df['Load']
Deflection_Two_mm_XY_Flat_df = Two_mm_XY_Flat_df['Deflection']

plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
plt.plot(Deflection_Two_mm_XY_Flat_df, Load_Two_mm_XY_Flat_df)
plt.xlabel('deflection (mm)')
plt.ylabel('load (N)')
plt.title('Load Deflection Curve of ABS 2mm Size in\n4-point bending flatwise test among XY alignment')
plt.legend(['2mm_XY_Flat'])

plt.subplot(1,2,2)
plt.plot(Strain_Two_mm_XY_Flat_df, Stress_Two_mm_XY_Flat_df)
plt.xlabel('strain (%)')
plt.ylabel('stress (MPa)')
plt.title('Stress Strain Curve of ABS 2mm Size in\n4-point bending flatwise test among XY alignment')
plt.legend(['2mm_XY_Flat'])

plt.tight_layout()
plt.savefig('./Downloads/2mm_XY_Flat.png')
plt.show()
```
